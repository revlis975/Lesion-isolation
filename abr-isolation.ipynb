## Appyling pre-processing on a small diverse subset of the original dataset

```python
from google.colab.patches import cv2_imshow
I = []

for i in range(1,19):
   I.append(io.imread('/content/drive/MyDrive/abr/abr'+str(i)+'.JPG'))

#BGR TO RGB 

RGB = []
for i in range(0,17):
  RGB.append(cv2.cvtColor(I[i], cv2.COLOR_BGR2RGB))

image = cv2.cvtColor(I[1], cv2.COLOR_BGR2RGB)

#RGB TO BGR

bgrimage = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

#BGR TO HSV

HSV = []

for i in range(0,17):
  HSV.append(cv2.cvtColor(RGB[i], cv2.COLOR_BGR2HSV))

hsvImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(image)
cv2_imshow(bgrimage)
cv2_imshow(hsvImage)
cv2.waitKey(0)
```

## Splitting the H channel of the HSV colour space

```python

h = hsvImage[:,:,0]

cv2_imshow(h)

h = []

for i in range(0,17):
  h.append(HSV[i][:,:, 0])
```

### Applying the thresholded image as mask on the original leaf to get the isolated infection

```python
masked = []

for i in range(0,17):
  threshold,thresh = cv2.threshold(h[i], 30, 255, cv2.THRESH_BINARY_INV)
  masked.append(cv2.bitwise_and(RGB[i],RGB[i],mask=thresh))
  cv2_imshow(RGB[i])
  cv2_imshow(masked[i])
```
